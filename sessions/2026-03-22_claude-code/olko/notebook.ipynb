## What is Claude Code? (3 min)

**Claude Code** is an **agentic AI coding tool** built by Anthropic.

It lives in your terminal (or IDE) and can:

| Capability | Description |
|---|---|
| 🔍 Read codebase | Understands project structure, conventions, dependencies |
| ✏️ Edit files | Creates, modifies, and refactors code directly |
| 💻 Run commands | Executes shell commands, tests, linters, builds |
| 🔗 Integrate | Git, GitHub, Jira, Sentry, Notion, MCP tools, and more |
| 🧠 Remember | Persistent memory via CLAUDE.md, skills, agents |

### Where can you use it?

| Platform | How |
|---|---|
| **Terminal CLI** | `claude` — the primary interface |
| **VS Code / Cursor** | Claude Code extension |
| **JetBrains IDEs** | Plugin (IntelliJ, PyCharm, etc.) |
| **Desktop App** | Standalone Electron app |
| **Web** | [claude.ai/code](https://claude.ai/code) |
| **Mobile** | Remote Control — monitor & approve from phone |

## Installation & First Steps (2 min)

### Install

```bash
# macOS / Linux — native installer
curl -fsSL https://claude.ai/install.sh | bash

# macOS — Homebrew
brew install --cask claude-code

# Windows — PowerShell
irm https://claude.ai/install.ps1 | iex
```

### Start

```bash
cd your-project
claude            # interactive session
```

On first launch you'll log in with a **Claude subscription** or **Anthropic Console** account.

### Quick one-shot

```bash
claude "explain the auth flow in this repo"
```

## CLAUDE.md — Persistent Memory (3 min)

CLAUDE.md is the **most important customization** you can make. It tells Claude your project's conventions, architecture, and workflow rules.

### File Locations

| Scope | Path | Purpose |
|---|---|---|
| **Project** | `./CLAUDE.md` | Repo-specific rules (committed to git) |
| **Personal** | `~/.claude/CLAUDE.md` | Your personal preferences (global) |
| **Org-wide** | System admin path | Organization defaults |
| **Modular rules** | `.claude/rules/*.md` | Path-scoped rules with frontmatter |

### Quick Start

```
/init          # Claude analyzes your project and generates CLAUDE.md
```

### Auto Memory

Claude also writes notes to itself in:

```
~/.claude/projects/<project-hash>/memory/
```

These are learned preferences — e.g., "user prefers tabs" — that persist across sessions.

### What to Include vs Exclude

| ✅ Include | ❌ Exclude |
|---|---|
| Code style rules (formatting, naming) | Obvious language features |
| Build / lint / test commands | Entire API docs (use `@file` instead) |
| Architecture overview (key directories) | Secrets or credentials |
| Workflow rules (PR process, branching) | Very long lists (keep < 200 lines) |
| Framework & library choices | Duplicate info already in README |
| Common gotchas & pitfalls | |

### Modular Rules (`.claude/rules/`)

```yaml
# .claude/rules/api-style.md
---
paths:
  - src/api/**
---
- All API handlers must validate input with Pydantic
- Return standard error envelope: {"error": {"code": ..., "message": ...}}
```

### Example CLAUDE.md

```markdown
# Code style
- Use ES modules (import/export), not CommonJS
- 2-space indentation, TypeScript preferred

# Workflow
- Run `npm run lint` after editing
- Run `npm test -- --related` before committing

# Architecture
- API handlers in `src/api/handlers/`
- Shared utils in `src/lib/`
```

## Skills — Reusable Workflows (3 min)

Skills are **reusable prompt templates** that package multi-step workflows.

### Location

| Scope | Path |
|---|---|
| Personal | `~/.claude/skills/<name>/SKILL.md` |
| Project | `.claude/skills/<name>/SKILL.md` |

### YAML Frontmatter Fields

| Field | Purpose |
|---|---|
| `name` | Skill name (also used as `/command`) |
| `description` | When Claude should auto-load this skill |
| `disable-model-invocation` | Only run when explicitly called |
| `allowed-tools` | Restrict which tools the skill can use |
| `context` | `fork` = run in isolated subagent |
| `agent` | Use a specific agent for execution |

### Arguments

- `$ARGUMENTS` — full argument string
- `$0`, `$1`, … — positional arguments

### Invocation

- `/skill-name arg1 arg2` — explicit
- Automatic — Claude loads matching skill when relevant

### Bundled Skills

`/batch`, `/claude-api`, `/debug`, `/loop`, `/simplify`

### Advanced Features

- **Bundled files**: Skills can include supporting templates, scripts, and examples alongside `SKILL.md`
- **Dynamic context**: `` !`shell command` `` runs a shell command and injects output into the prompt before execution
- **`context: fork`**: Run in an isolated subagent (separate context window)

### Example Skill

```yaml
# .claude/skills/fix-issue/SKILL.md
---
name: fix-issue
description: Fix a GitHub issue by number
disable-model-invocation: true
---
Fix GitHub issue $ARGUMENTS.

1. Read issue with `gh issue view $0`
2. Search codebase for relevant files
3. Implement fix
4. Write tests, run them
5. Lint and typecheck
6. Commit and create PR
```

Usage:

```
/fix-issue 42
```

## Agents (.claude/agents/) — Custom Subagents (2 min)

Agents are **specialized Claude instances** that run in their own context window — they don't clutter your main conversation.

### Location

| Scope | Path |
|---|---|
| Project | `.claude/agents/<name>.md` |
| Personal | `~/.claude/agents/<name>.md` |

### Built-in Agents

| Agent | Purpose |
|---|---|
| **Explore** | Fast, read-only analysis (uses Haiku) |
| **Plan** | Deep research for implementation planning |

### Frontmatter Fields

| Field | Purpose |
|---|---|
| `name` | Agent name |
| `description` | When to use (Claude reads this) |
| `tools` | Allowed tools (e.g., `Read, Grep, Glob, Bash`) |
| `model` | `opus`, `sonnet`, `haiku` |
| `permissionMode` | `plan`, `auto-accept`, etc. |
| `mcpServers` | MCP servers available to agent |
| `hooks` | Agent-specific hooks |
| `memory` | `user`, `project`, or `local` |
| `isolation` | `worktree` for git isolation |

### How to Use Agents

```bash
# List / create / manage
/agents

# Invoke in conversation
@security-reviewer check the latest PR

# Run as full-session agent
claude --agent code-reviewer

# Background a running task
Ctrl+B
```

### Example Agent

```yaml
# .claude/agents/security-reviewer.md
---
name: security-reviewer
description: Reviews code for security vulnerabilities. Use proactively after code changes.
tools: Read, Grep, Glob, Bash
model: sonnet
memory: project
---
You are a senior security engineer. Review code for:

- Injection vulnerabilities (SQL, XSS, command injection)
- Authentication and authorization flaws
- Secrets or credentials in code
- Insecure data handling

Provide specific line references and suggested fixes.
```

## Hooks — Deterministic Automation (2 min)

Hooks are **guaranteed actions** that fire at specific lifecycle points.

> Unlike CLAUDE.md instructions (which are advisory), hooks **always execute**.

### Lifecycle Events

| Event | When it fires |
|---|---|
| `SessionStart` | New session begins |
| `PreToolUse` | Before a tool runs (can **block** with exit code 2) |
| `PostToolUse` | After a tool completes |
| `Stop` | Claude finishes its turn |
| `Notification` | Claude wants your attention |
| `SubagentStart` / `SubagentStop` | Subagent lifecycle |

### Hook Types

| Type | Description |
|---|---|
| `command` | Shell command |
| `http` | HTTP endpoint call |
| `prompt` | LLM prompt (ask Claude) |
| `agent` | Invoke an agent |

### Matchers

Regex on tool name: `"Edit\|Write"`, `"Bash"`, `"mcp__.*"`

Manage with: `/hooks`

### Example: Auto-lint after every file edit

```json
{
  "hooks": {
    "PostToolUse": [{
      "matcher": "Edit|Write",
      "hooks": [{"type": "command", "command": "npm run lint --fix"}]
    }]
  }
}
```

### Example: Desktop notification when Claude needs attention

```json
{
  "hooks": {
    "Notification": [{
      "matcher": "",
      "hooks": [{
        "type": "command",
        "command": "osascript -e 'display notification \"Claude needs you\" with title \"Claude Code\"'"
      }]
    }]
  }
}
```

## MCP — Connect External Tools (2 min)

**Model Context Protocol** is an open standard that lets Claude Code talk to external services.

### What You Can Connect

GitHub · Sentry · Notion · Figma · Jira · Slack · databases · your custom tools

### Transport Types

| Type | Best for |
|---|---|
| **HTTP** (recommended) | Cloud services, OAuth |
| **SSE** | Server-sent events |
| **stdio** | Local processes |

### Adding MCP Servers

```bash
# GitHub
claude mcp add --transport http github https://api.githubcopilot.com/mcp/

# Sentry
claude mcp add --transport http sentry https://mcp.sentry.dev/mcp

# Database (local)
claude mcp add --transport stdio db -- npx -y @bytebase/dbhub --dsn "postgresql://..."
```

### Key Features

| Feature | Detail |
|---|---|
| **`/mcp`** | Manage & authenticate connections |
| **OAuth 2.0** | Secure authentication flow |
| **Scopes** | `local` (default), `project` (`.mcp.json`, shared via git), `user` (cross-project) |
| **Resources** | `@github:issue://123` — reference MCP resources |
| **Serve mode** | `claude mcp serve` — use Claude Code **as** an MCP server |

## Best Practices & Productivity (2 min)

### Golden Rule

> **Give Claude a way to VERIFY its own work** — tests, screenshots, expected outputs, linters.

### Recommended Workflow

```
Explore → Plan → Implement → Commit
```

| Step | How |
|---|---|
| 1. **Plan** | `Shift+Tab` (Plan Mode) — read-only analysis |
| 2. **Refine** | `Ctrl+G` — edit the plan in your editor |
| 3. **Implement** | `Shift+Tab` (Normal Mode) — let Claude code |
| 4. **Ship** | Commit, create PR |

### Context Management

| Technique | When |
|---|---|
| `/clear` | Between unrelated tasks |
| `/compact` | When context fills up |
| Subagents | For investigation (separate context) |
| `Esc` | Stop mid-action |
| `Esc` `Esc` | Rewind menu |
| `/btw` | Side questions without growing context |

### Common Pitfalls

| Pitfall | Fix |
|---|---|
| Kitchen-sink session (too many topics) | `/clear` between tasks |
| Repeated corrections (Claude keeps getting it wrong) | `/clear` + better prompt after 2 failures |
| CLAUDE.md too long | Prune to under 200 lines |
| No verification step | Always provide tests or expected output |
| Infinite exploration | Scope narrowly or use subagents |

### Multi-Session Patterns

| Pattern | How |
|---|---|
| **Writer / Reviewer** | Session A writes code, Session B reviews |
| **Isolated worktree** | `--worktree feature-x` |
| **CI/CD** | `claude -p "query"` in pipelines |
| **Fan out** | `for file in $(cat files.txt); do claude -p "Migrate $file"; done` |
| **Pipe** | `cat error.log \| claude -p "explain root cause" > fix.txt` |

## LIVE DEMO — Build a Real Feature with Claude Code (~5 min)

**Scenario: Build a FastAPI health-check endpoint from scratch**

---

### Step 1 — Setup project

```bash
mkdir /tmp/demo-api && cd /tmp/demo-api
git init
claude
```

### Step 2 — Generate CLAUDE.md

```
/init
```

> Claude analyzes the (empty) project. Briefly review what it generated.

### Step 3 — Ask Claude to scaffold a FastAPI project

```
Create a FastAPI application with:
- A health check endpoint at GET /health that returns {"status": "ok", "timestamp": "..."}
- A GET /items endpoint that returns a list of sample items
- Include proper type hints and Pydantic models
- Add requirements.txt with fastapi and uvicorn
- Add a test file using pytest
```

> 👀 Watch Claude create files, install deps, write tests

### Step 4 — Show Plan Mode

```
Shift+Tab   (switch to Plan Mode)
```

```
How would you add authentication to this API? What files would change?
```

```
Shift+Tab   (back to Normal Mode)
```

### Step 5 — Run tests

```
Run the tests and fix any failures
```

### Step 6 — Use @-mentions

```
Look at @src/main.py — add request logging middleware
```

### Step 7 — Commit

```
Commit all changes with a descriptive message
```

### Step 8 — Non-interactive mode (separate terminal)

```bash
cat src/main.py | claude -p "review this code for security issues"
```

### Step 9 — Resume

```bash
claude --continue    # pick up where we left off
```
